```python
%load_ext autoreload
%reload_ext autoreload
import newspaper
import pandas as pd
import numpy as np
from tqdm import *
from newspaper import Article
from multiprocessing import Pool, cpu_count
from contextlib import closing
```

```python
df = pd.read_csv('./data/uci-news-aggregator.csv')
df = df[['TITLE', 'URL', 'HOSTNAME']]
df.reset_index(inplace=True)
df['ID_URL'] = df.apply(lambda x: (x["index"], x.URL), axis=1)
df = pd.concat([df, pd.Series(np.ones(len(df)), name='CLASS')], axis=1)
```

```python
df.sample(1)
```

```python
def get_body(tupple):
    index, url = tupple
    a = Article(url, language='en')
    a.download()
    a.parse()
    return (index, u''.join(a.text).encode('utf-8').strip())
```

```python
pool = Pool(max(2,cpu_count()-1))
from time import time
t0 = time()
with closing(pool) as p:
    corpus = p.map(get_body, df['ID_URL'].sample(10000).values)
print("it took {} seconds".format(str(int(time()) - t0)))
```
